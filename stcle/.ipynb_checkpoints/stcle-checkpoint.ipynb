```python
# -*- coding: utf-8 -*-

from datetime import datetime
import calendar

import tkinter as tk
from tkinter import ttk

from tkinter.tix import Tk, Control, ComboBox, Button, Balloon, Label, Entry, Text #升级的组合控件包

from tkinter.messagebox import showinfo, showwarning, showerror #各种类型的提示框
from tkinter.constants import *

import tkinter.font as tkFont
```

```python
import json

#如果没有.json文件，则创建 stcle字典
try:
    with open('stcle.json', 'r', encoding='UTF-8') as f:
        stcle = json.load(f)
except FileNotFoundError:
    stcle = {
        'basic_sciences':  [{'date':'3000', 'region':'天堂', 'event': '伊甸园', }, ],
        'key_technology':  [{'date':'3000', 'region':'天堂', 'event': '伊甸园', }, ],
        'bulk_commodity':  [{'date':'3000', 'region':'天堂', 'event': '伊甸园', }, ],
        'dayly_life':      [{'date':'3000', 'region':'天堂', 'event': '伊甸园', }, ],
        'population_size': [{'date':'3000', 'region':'天堂', 'event': '伊甸园', }, ],
        'macro_economy':   [{'date':'3000', 'region':'天堂', 'event': '伊甸园', }, ],
        }
```

```python
root = Tk() # 初始化Tk()  root便是你布局的根节点了，以后的布局都在它之上

root.title("科学、技术、商品、生活、经济") # 设置窗口标题

width = root.winfo_reqwidth() + 400 #窗口大小
height = root.winfo_reqheight() + 100 #窗口大小
x, y = (root.winfo_screenwidth()  - width )/2, (root.winfo_screenheight() - height)/2
#root.attributes("-toolwindow", 1)  #参数1，设置工具栏样式窗口。#参数1，没有最大化和最小化的按钮。
#root["background"] = "DarkBlue"
root["bd"] = 4
root["relief"] = GROOVE #RAISED
root.geometry('%dx%d+%d+%d' % (width, height, x, y ) ) #窗口位置居中,设置窗口大小 注意：是x 不是*

root.resizable(width=True, height=True) # 设置窗口是否可以变化长/宽，False不可变，True可变，默认为True

root.tk.eval('package require Tix') #引入升级包，这样才能使用升级的组合控件

#创建一个状态显示标签
info_status = Label(root, width=300, relief=GROOVE, font=("Arial",11), fg='red', bd=1, anchor = SW)
info_status.place(x=0,y=0, relx=0/16,rely=14/15, relwidth=8/8,relheigh=1/15,anchor=NW)

#以下是输入选择，发生年份和发生的事情
date_str = tk.StringVar()
region_str = tk.StringVar()
event_str = tk.StringVar()

#实现按钮选择
def sciences_button_call():
    if sciences_button.configure('relief')[-1] == RAISED:
        a_str.set('Sciences')
        sciences_button.configure(relief=SUNKEN, bg='Moccasin')
        technology_button.configure(relief=RAISED, bg='LightGreen')
        commodity_button.configure(relief=RAISED, bg='LightGreen')
        life_button.configure(relief=RAISED, bg='LightGreen')
        population_button.configure(relief=RAISED, bg='LightGreen')
        economy_button.configure(relief=RAISED, bg='LightGreen')
    else:
        sciences_button.configure(relief=RAISED, bg='LightGreen')
        a_str.set('')

def technology_button_call():
    if technology_button.configure('relief')[-1] == RAISED:
        a_str.set('Technology')
        sciences_button.configure(relief=RAISED, bg='LightGreen')
        technology_button.configure(relief=SUNKEN, bg='Moccasin')
        commodity_button.configure(relief=RAISED, bg='LightGreen')
        life_button.configure(relief=RAISED, bg='LightGreen')
        population_button.configure(relief=RAISED, bg='LightGreen')
        economy_button.configure(relief=RAISED, bg='LightGreen')
    else:
        technology_button.configure(relief=RAISED, bg='LightGreen')
        a_str.set('')
        
def commodity_button_call():
    if commodity_button.configure('relief')[-1] == RAISED:
        a_str.set('Commodity')
        sciences_button.configure(relief=RAISED, bg='LightGreen')
        technology_button.configure(relief=RAISED, bg='LightGreen')
        commodity_button.configure(relief=SUNKEN, bg='Moccasin')
        life_button.configure(relief=RAISED, bg='LightGreen')
        population_button.configure(relief=RAISED, bg='LightGreen')
        economy_button.configure(relief=RAISED, bg='LightGreen')
    else:
        commodity_button.configure(relief=RAISED, bg='LightGreen')
        a_str.set('')

def life_button_call():
    if life_button.configure('relief')[-1] == RAISED:
        a_str.set('Life')
        sciences_button.configure(relief=RAISED, bg='LightGreen')
        technology_button.configure(relief=RAISED, bg='LightGreen')
        commodity_button.configure(relief=RAISED, bg='LightGreen')
        life_button.configure(relief=SUNKEN, bg='Moccasin')
        population_button.configure(relief=RAISED, bg='LightGreen')
        economy_button.configure(relief=RAISED, bg='LightGreen')
    else:
        life_button.configure(relief=RAISED, bg='LightGreen')
        a_str.set('')
        
def population_button_call():
    if population_button.configure('relief')[-1] == RAISED:
        a_str.set('Population')
        sciences_button.configure(relief=RAISED, bg='LightGreen')
        technology_button.configure(relief=RAISED, bg='LightGreen')
        commodity_button.configure(relief=RAISED, bg='LightGreen')
        life_button.configure(relief=RAISED, bg='LightGreen')
        population_button.configure(relief=SUNKEN, bg='Moccasin')
        economy_button.configure(relief=RAISED, bg='LightGreen')
    else:
        population_button.configure(relief=RAISED, bg='LightGreen')
        a_str.set('')
        
def economy_button_call():
    if economy_button.configure('relief')[-1] == RAISED:
        a_str.set('Economy')
        sciences_button.configure(relief=RAISED, bg='LightGreen')
        technology_button.configure(relief=RAISED, bg='LightGreen')
        commodity_button.configure(relief=RAISED, bg='LightGreen')
        life_button.configure(relief=RAISED, bg='LightGreen')
        population_button.configure(relief=RAISED, bg='LightGreen')
        economy_button.configure(relief=SUNKEN, bg='Moccasin')
    else:
        economy_button.configure(relief=RAISED, bg='LightGreen')
        a_str.set('')
        

#显示 S T C L P E的标签
a_str = tk.StringVar()
a_str.set('')

a_lable = Label(root, textvariable=a_str, font=("Fixdsys",25,tkFont.BOLD), fg="black")
a_lable.place(x=0,y=0, relx=1/2, rely=1/4, relwidth=1/2, relheigh=1/3, anchor=NW)

#显示 日期、地区、事件的输入框
date_lable = Label(root, text="日 期", bg="Moccasin", bd=1, font=("Arial",11), relief=RIDGE)
date_lable.place(x=0,y=0, relx=1/16, rely=6/15, relwidth=1/8, relheigh=1/15, anchor=NW)
date = Entry(root, textvariable = date_str, bg="Wheat", bd=2, relief=SUNKEN
            ).place(x=0,y=0, relx=3/16, rely=6/15, relwidth=2/8, relheigh=1/15, anchor=NW)

region_lable = Label(root, text="地 区", bg="Moccasin", bd=1, font=("Arial",11), relief=RIDGE)
region_lable.place(x=0,y=0, relx=1/16, rely=8/15, relwidth=1/8, relheigh=1/15, anchor=NW)
region = Entry(root, textvariable = region_str, bg="Wheat", bd=2, relief=SUNKEN
            ).place(x=0,y=0, relx=3/16, rely=8/15, relwidth=2/8, relheigh=1/15, anchor=NW)

event_lable = Label(root, text="事 件", bg="Moccasin", bd=1, font=("Arial",11), relief=RIDGE)
event_lable.place(x=0,y=0, relx=1/16, rely=10/15, relwidth=1/8, relheigh=1/15, anchor=NW)
event = Entry(root, textvariable = event_str, bg="Wheat", bd=2, relief=SUNKEN
            ).place(x=0,y=0, relx=3/16, rely=10/15, relwidth=5/8, relheigh=1/15, anchor=NW)

sciences_button = Button(root, text="科学原理", bg="LightGreen",bd=2, font=("Arial",12), relief=RAISED,
                        command = sciences_button_call)
sciences_button.place(x=0,y=0, relx=10/48, rely=2/15, relwidth=1/8, relheigh=1/10, anchor=NW)
sciences_balloon = Balloon(root, statusbar=info_status)
sciences_balloon.bind_widget(sciences_button, balloonmsg='', statusmsg='说明：点按这个按钮输入发生的“科学原理”事件。')

technology_button = Button(root, text="关键技术", bg="LightGreen",bd=2, font=("Arial",12), relief=RAISED,
                          command = technology_button_call)
technology_button.place(x=0,y=0, relx=16/48, rely=2/15, relwidth=1/8, relheigh=1/10, anchor=NW)
technology_balloon = Balloon(root, statusbar=info_status)
technology_balloon.bind_widget(technology_button, balloonmsg='', statusmsg='说明：点按这个按钮输入发生的“关键技术”事件。')


commodity_button = Button(root, text="大宗商品", bg="LightGreen",bd=2, font=("Arial",12), relief=RAISED,
                         command = commodity_button_call)
commodity_button.place(x=0,y=0, relx=22/48, rely=2/15, relwidth=1/8, relheigh=1/10, anchor=NW)
commodity_balloon = Balloon(root, statusbar=info_status)
commodity_balloon.bind_widget(commodity_button, balloonmsg='', statusmsg='说明：点按这个按钮输入发生的“大宗商品”事件。')

life_button = Button(root, text="日常生活", bg="LightGreen",bd=2, font=("Arial",12), relief=RAISED,
                    command = life_button_call)
life_button.place(x=0,y=0, relx=28/48, rely=2/15, relwidth=1/8, relheigh=1/10, anchor=NW)
life_balloon = Balloon(root, statusbar=info_status)
life_balloon.bind_widget(life_button, balloonmsg='', statusmsg='说明：点按这个按钮输入发生的“日常生活”事件。')

population_button = Button(root, text="人口数量", bg="LightGreen",bd=2, font=("Arial",12), relief=RAISED,
                          command = population_button_call)
population_button.place(x=0,y=0, relx=34/48, rely=2/15, relwidth=1/8, relheigh=1/10, anchor=NW)
population_balloon = Balloon(root, statusbar=info_status)
population_balloon.bind_widget(population_button, balloonmsg='', statusmsg='说明：点按这个按钮输入发生的“人口数量”事件。')

economy_button = Button(root, text="宏观经济", bg="LightGreen",bd=2, font=("Arial",12), relief=RAISED,
                       command = economy_button_call)
economy_button.place(x=0,y=0, relx=40/48, rely=2/15, relwidth=1/8, relheigh=1/10, anchor=NW)
economy_balloon = Balloon(root, statusbar=info_status)
economy_balloon.bind_widget(economy_button, balloonmsg='', statusmsg='说明：点按这个按钮输入发生的“宏观经济”事件。')


#保存设定的数据到文件

def add_json():
    date = date_str.get()
    region = region_str.get()
    event = event_str.get()
    stcle_instance = {'date':date, 'region':region, 'event':event}
    if sciences_button.configure('relief')[-1] == SUNKEN:
        stcle["basic_sciences"].append(stcle_instance)
    if technology_button.configure('relief')[-1] == SUNKEN:
        stcle["key_technology"].append(stcle_instance)
    if commodity_button.configure('relief')[-1] == SUNKEN:
        stcle["bulk_commodity"].append(stcle_instance)
    if life_button.configure('relief')[-1] == SUNKEN:
        stcle["dayly_life"].append(stcle_instance)
    if population_button.configure('relief')[-1] == SUNKEN:
        stcle["population_size"].append(stcle_instance)
    if economy_button.configure('relief')[-1] == SUNKEN:
        stcle["macro_economy"].append(stcle_instance)
    date_str.set('')
    region_str.set('')
    event_str.set('')
    
def save_quit_json():
    with open('stcle.json', 'w', encoding='utf-8') as f:
        json.dump(stcle, f)
    root.destroy()

    
add_button = Button(root, text = '增加记录', command = add_json, 
                     relief = RAISED, font = ("Arial",10), bd = 3
                    ).place(x=0,y=0, relx=5/15,rely=13/16, relwidth=1/8,relheigh=1/15)    

save_quit_button = Button(root, text = '保存退出', command = save_quit_json, 
                     relief = RAISED, font = ("Arial",10), bd = 3
                    ).place(x=0,y=0, relx=8/15,rely=13/16, relwidth=1/8,relheigh=1/15)  

root.mainloop()
```
