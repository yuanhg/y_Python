```python
root = Tk() # 初始化Tk()  root便是你布局的根节点了，以后的布局都在它之上

root.title("科学、技术、商品、生活、经济") # 设置窗口标题

width = root.winfo_reqwidth() + 400 #窗口大小
height = root.winfo_reqheight() + 100 #窗口大小
x, y = (root.winfo_screenwidth()  - width )/2, (root.winfo_screenheight() - height)/2
#root.attributes("-toolwindow", 1)  #参数1，设置工具栏样式窗口。#参数1，没有最大化和最小化的按钮。
#root["background"] = "DarkBlue"
root["bd"] = 4
root["relief"] = GROOVE #RAISED
root.geometry('%dx%d+%d+%d' % (width, height, x, y ) ) #窗口位置居中,设置窗口大小 注意：是x 不是*

root.resizable(width=True, height=True) # 设置窗口是否可以变化长/宽，False不可变，True可变，默认为True

root.tk.eval('package require Tix') #引入升级包，这样才能使用升级的组合控件

#创建一个状态显示标签
info_status = Label(root, width=300, relief=GROOVE, font=("Arial",13), fg='red', bd=1, anchor = SW)
info_status.place(x=0,y=0, relx=0/20,rely=7/8, relwidth=20/20,relheigh=3/30,anchor=NW)

#以下是输入选择，发生年份和发生的事情
date_str = tk.StringVar()

date_lable = Label(root, text="发生年份", bg="white", bd=2, font=("Arial",12), relief=RIDGE)
date_lable.place(x=0,y=0, relx=1/16, rely=3/15, relwidth=1/8, relheigh=2/15, anchor=NW)

date = Entry(root, textvariable = date_str, bg="LightCoral", bd=2, relief=SUNKEN
            ).place(x=0,y=0, relx=1/16, rely=6/15, relwidth=1/8, relheigh=2/15, anchor=NW)

sciences_button = Button(root, text="科学原理", bg="white",bd=2, font=("Arial",12), relief=RAISED)
sciences_button.place(x=0,y=0, relx=10/48, rely=3/15, relwidth=1/8, relheigh=2/15, anchor=NW)
sciences_balloon = Balloon(root, statusbar=info_status)
sciences_balloon.bind_widget(sciences_button, balloonmsg='', statusmsg='说明：点按这个按钮输入发生的“科学原理”事件。')

technology_button = Button(root, text="关键技术", bg="white",bd=2, font=("Arial",12), relief=RAISED)
technology_button.place(x=0,y=0, relx=16/48, rely=3/15, relwidth=1/8, relheigh=2/15, anchor=NW)
technology_balloon = Balloon(root, statusbar=info_status)
technology_balloon.bind_widget(technology_button, balloonmsg='', statusmsg='说明：点按这个按钮输入发生的“关键技术”事件。')


commodity_button = Button(root, text="大宗商品", bg="white",bd=2, font=("Arial",12), relief=RAISED)
commodity_button.place(x=0,y=0, relx=22/48, rely=3/15, relwidth=1/8, relheigh=2/15, anchor=NW)
commodity_balloon = Balloon(root, statusbar=info_status)
commodity_balloon.bind_widget(commodity_button, balloonmsg='', statusmsg='说明：点按这个按钮输入发生的“大宗商品”事件。')

life_button = Button(root, text="日常生活", bg="white",bd=2, font=("Arial",12), relief=RAISED)
life_button.place(x=0,y=0, relx=28/48, rely=3/15, relwidth=1/8, relheigh=2/15, anchor=NW)
life_balloon = Balloon(root, statusbar=info_status)
life_balloon.bind_widget(life_button, balloonmsg='', statusmsg='说明：点按这个按钮输入发生的“日常生活”事件。')

population_button = Button(root, text="人口数量", bg="white",bd=2, font=("Arial",12), relief=RAISED)
population_button.place(x=0,y=0, relx=34/48, rely=3/15, relwidth=1/8, relheigh=2/15, anchor=NW)
population_balloon = Balloon(root, statusbar=info_status)
population_balloon.bind_widget(population_button, balloonmsg='', statusmsg='说明：点按这个按钮输入发生的“人口数量”事件。')

economy_button = Button(root, text="宏观经济", bg="white",bd=2, font=("Arial",12), relief=RAISED)
economy_button.place(x=0,y=0, relx=40/48, rely=3/15, relwidth=1/8, relheigh=2/15, anchor=NW)
economy_balloon = Balloon(root, statusbar=info_status)
economy_balloon.bind_widget(economy_button, balloonmsg='', statusmsg='说明：点按这个按钮输入发生的“宏观经济”事件。')

root.mainloop()
```
