```python
"""[]
1.Bubble Sort


  
  (n^2 (n (n^2
(1)

In-place


  
  
  
  
"""

def bubble_sort(arr):
    for i in range(1, len(arr)):
        for j in range(0, len(arr)-i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  
    return arr


numbers = np.random.randint(100000000,size=100000)
print(numbers)
%time bubble_sort(numbers)
print(numbers)
```

```python
"""[]
2.(Quick Sort) n  (nlogn) 
 (n^2)  (nlogn) 
(nlogn)  O(nlogn) 
inner loop

Divide and conquerlistsub-lists

 
  (nlogn) (nlogn) (n^2
(nlogn)

In-place


   pivot;
  
  partition
  recursive

  
  iteration
"""

#
def quick_sort(arr, left=None, right=None):
    
    left = 0 if not isinstance(left,(int, float)) else left
    right = len(arr)-1 if not isinstance(right,(int, float)) else right
    
    if left < right:
        partitionIndex = partition(arr, left, right)
        quick_sort_2(arr, left, partitionIndex-1)
        quick_sort_2(arr, partitionIndex+1, right)
        
    return arr

#
def partition(arr,left,right):
    #
    mid = arr[left]
    leftmark = left+1
    rightmark = right
    done = False
    
    while not done:
        while leftmark<=rightmark and arr[leftmark]<=mid:
            leftmark=leftmark+1
        while rightmark>=leftmark and arr[rightmark]>=mid:
            rightmark-=1
        #
        if rightmark<leftmark:  
            done=True
        #
        else:
            arr[leftmark], arr[rightmark] = arr[rightmark], arr[leftmark]
            
    arr[left], arr[rightmark] = arr[rightmark], arr[left]
    
    return rightmark

def swap(arr, i, j):
    arr[i], arr[j] = arr[j], arr[i]


#.
#quick_sort_1 = lambda array: array if len(array) <= 1 else quick_sort_1([item for item in array[1:] if item <= array[0]]) + [array[0]] + quick_sort_1([item for item in array[1:] if item > array[0]])
quick_sort_1 = lambda array: array if len(array) <= 1 else (
    quick_sort_1([item for item in array[1:] if item <= array[0]])
    + [array[0]]
    + quick_sort_1([item for item in array[1:] if item > array[0]]) )

#
def quick_sort_2(array, l, r):
    if l >= r:
        return
    stack = []
    stack.append(l)
    stack.append(r)
    while stack:
        low = stack.pop(0)
        high = stack.pop(0)
        if high - low <= 0:
            continue
        x = array[high]
        i = low - 1
        for j in range(low, high):
            if array[j] <= x:
                i += 1
                array[i], array[j] = array[j], array[i]
        array[i + 1], array[high] = array[high], array[i + 1]
        stack.extend([low, i, i + 2, high])

        
numbers = np.random.randint(100000000,size=1000000) #
print(numbers)
np.random.shuffle(numbers)
print(numbers)
%time numbers = quick_sort(numbers)
#%time quick_sort_1(numbers)
#%time quick_sort_2(numbers, 1, len(numbers)-1)
print(numbers)
```

```python
"""[]
1.(Insertion Sort)




 (Shell Sort)

O(N^2)
O(1)

In-place


  
  
  
"""
def insertion_sort(arr):
    for i in range(len(arr)):
        preIndex = i-1
        current = arr[i]
        while preIndex >= 0 and arr[preIndex] > current:
            arr[preIndex+1] = arr[preIndex]
            preIndex-=1
        arr[preIndex+1] = current
    return arr

numbers = np.random.randint(100000000,size=100000)
print(numbers)
%time insertion_sort(numbers)
print(numbers)
```

```python
# 
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation
%matplotlib auto  

N = 15
MAX = 100
MIN = 10

y1 = np.random.randint(MIN, MAX, size=N)
y3 = np.zeros(N, dtype = np.int) #y1
y5 = np.zeros(N, dtype = np.int) #y1 1
y2 = np.zeros(N, dtype = np.int) #


figsize = 9, 5
fig, ax = plt.subplots(figsize = figsize)

plt.title('insertion Sort', fontsize=24, color='DodgerBlue')
plt.xticks([]) #x
plt.box(False)

bar1 = ax.bar(range(N), y1, color='DarkSeaGreen') #
bar3 = ax.bar(range(N), y3, color='DarkGoldenrod') #
bar5 = ax.bar(range(N), y5, color='Crimson')  #
bar2 = ax.bar(range(N), y2, color='DarkGoldenrod')


def init():    
    ax.set_ylim(-MAX, MAX)
    ax.grid(which='major', axis='y', linestyle='-', color='DarkSeaGreen') 
    #plt.yticks([0],[])      
    #return bar1, bar2


def insertion_sort_frames():
    
    for i in range(len(y1)):

        preIndex = i-1
        current = y1[i]
        
        y3[i] = y1[i]
        if preIndex >= 0:
            y5[preIndex] = y1[preIndex] #1
        yield i
        
        while preIndex >= 0 and y1[preIndex] > current:
            if preIndex + 1 == i:
                y2[preIndex+1] = -y1[preIndex+1] 
                y1[preIndex+1] = 0
                y3[preIndex+1] = 0
                yield i
                               
            y1[preIndex+1] = y1[preIndex]
            y5[preIndex+1] = y1[preIndex] if preIndex + 1 == i else 0
            y1[preIndex] = 0
            y3[preIndex] = 0
            y5[preIndex] = 0
            y2[preIndex] = y2[preIndex+1]
            y2[preIndex+1] = 0
            if preIndex > 0 :
                y3[preIndex-1] = y1[preIndex-1]
            yield i
            
            preIndex-=1
            
        y1[preIndex+1] = current
        y3[preIndex+1] = 0
        y5[preIndex+1] = 0
        y2[preIndex+1] = 0 
        if preIndex >= 0:
            y3[preIndex] = 0 
            y5[preIndex] = 0
        yield i
    
    y3[i] = 0
    y5[i] = 0
    yield i



def update_bar(index):
    
    for rect ,y in zip(bar1, y1): rect.set_height(y)        
    for rect ,y in zip(bar3, y3): rect.set_height(y)  
    for rect ,y in zip(bar5, y5): rect.set_height(y) 
    for rect ,y in zip(bar2, y2): rect.set_height(y)    


ani = animation.FuncAnimation(fig, func=update_bar, frames=insertion_sort_frames(), init_func=init(),
                              interval = 600,  blit=False)

plt.show()
```

```python
"""[]
1.(Selection Sort) O(n) 


O(N^2)
O(1)

In-place


  
  
  
"""
def selection_sort(arr):
    for i in range(len(arr) - 1):
        # 
        minIndex = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[minIndex]: minIndex = j   #minIndes

        # arr[i]   arr[minIndex] 
        arr[i], arr[minIndex] = arr[minIndex], arr[i] 
            
    return arr

numbers = np.random.randint(100000000,size=100000)
print(numbers)
%time selection_sort(numbers)
print(numbers)
```

```python
"""[]
2.Heap Sort



  
  

  (N*logN)
 (1)

In-place


   H[0n-1]
  
   1 shift_down(0)
   2 1
"""

def buildMaxHeap(arr):
    import math
    for i in range(math.floor(len(arr)/2),-1,-1):
        heapify(arr,i)

def heapify(arr, i):
    left = 2*i+1
    right = 2*i+2
    largest = i
    if left < arrLen and arr[left] > arr[largest]:
        largest = left
    if right < arrLen and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        swap(arr, i, largest)
        heapify(arr, largest)

def swap(arr, i, j):
    arr[i], arr[j] = arr[j], arr[i]

def heap_sort(arr):
    global arrLen
    arrLen = len(arr)
    buildMaxHeap(arr)
    for i in range(len(arr)-1,0,-1):
        swap(arr,0,i)
        arrLen -=1
        heapify(arr, 0)
    return arr


numbers = np.random.randint(100000000,size=1000000) #
print(numbers)
np.random.shuffle(numbers)
print(numbers)
%time heap_sort(numbers)
print(numbers)
```

```python
"""[]
1.Merge sortDivide and Conquer

   2 :  

 O(nlogn) 


  
  
  
   3 
  
"""
def merge_sort(arr):
    import math
    if(len(arr)<2):
        return arr
    middle = math.floor(len(arr)/2)
    left, right = arr[0:middle], arr[middle:]
    return merge(merge_sort(left), merge_sort(right))

def merge(left,right):
    result = []
    while left and right:
        if left[0] <= right[0]:
            result.append(left.pop(0));
        else:
            result.append(right.pop(0));
    while left:
        result.append(left.pop(0));
    while right:
        result.append(right.pop(0));
    return result


numbers = np.random.randint(100000000,size=1000000) #
print(numbers)
np.random.shuffle(numbers)
print(numbers)
%time merge_sort(list(numbers))
print(numbers)
```

```python
"""[]
1.(Counting Sort)

"""
def counting_sort(arr, maxValue):
    bucketLen = maxValue+1
    bucket = [0]*bucketLen
    sortedIndex =0
    arrLen = len(arr)
    for i in range(arrLen):
        if not bucket[arr[i]]:
            bucket[arr[i]]=0
        bucket[arr[i]]+=1
    for j in range(bucketLen):
        while bucket[j]>0:
            arr[sortedIndex] = j
            sortedIndex+=1
            bucket[j]-=1
    return arr


numbers = np.random.randint(100000000,size=1000000) #
print(numbers)
np.random.shuffle(numbers)
print(numbers)
%time counting_sort(numbers)
print(numbers)
```

```python
"""
9.(Bucket Sort)

  
   N  K 
  

: 
: 
"""
def bucket_sort(s):
    """"""
    min_num = min(s)
    max_num = max(s)
    # 
    bucket_range = (max_num-min_num) / len(s)
    # 
    count_list = [ [] for i in range(len(s) + 1)]
    # 
    for i in s:
        count_list[int((i-min_num)//bucket_range)].append(i)
    s.clear()
    # sorted
    for i in count_list:
        for j in sorted(i):
            s.append(j)


numbers = np.random.randint(100000000,size=100000)
print(numbers)
%time bucket_sort(numbers)
print(numbers)
```
